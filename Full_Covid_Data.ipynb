# Data Analysis and Visualization of COVID 19

```python
'''This project involves the use of pandas and plotly in order to analyze COVID data and make visualizations
and analysis of this. In order to run these cells simply hit the run button. It is important to keep in mind 
that there are lots of 0 values in the dataframe. The reason for this is because many of the countries did not 
report numbers for specific catagories. Instead of keeping the values as NaN, I decided to make them 0 so it would
be easier for visualization. Please open this in a google colab, as visualizations are not working on here''' 
import pandas as pd
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.express as px
import plotly.figure_factory as ff
from plotly import tools

#Reading in the CSV data and creating headers
df = pd.read_csv(
    "/content/covid_data.csv",
    names=[
        "Iso Code",
        "Continent",
        "Location",
        "Date",
        "Total Cases",
        "New Cases",
        "New Cases Smoothed",
        "Total Deaths",
        "New Deaths",
        "New Deaths Smoothed",
        "Total Cases per Million",
        "New Cases Per Million",
        "New Cases Smoothed Per Million",
        "Total Deaths Per Million",
        "New Deaths Per Million",
        "New Deaths Smoothed Per Million",
        "Reproduction Rate",
        "ICU Patients",
        "ICU Patients Per Million",
        "Hospital Patients",
        "Hospital Patients Per Million",
        "Weekly ICU Admissions",
        "Weekly ICU Admissions Per Million",
        "Weekly Hospital Admissions",
        "Weekly Hospital Admissions Per Million",
        "New Tests",
        "Total Tests",
        "Total Tests Per Thousand",
        "New Tests Per Thousand",
        "New Tests Smoothed",
        "New Tests Smoothed Per Thousand",
        "Positive Rate",
        "Tests Per Case",
        "Tests Units",
        "Total Vaccinations",
        "People Vaccinated",
        "People Fully Vaccinated",
        "New Vaccinations",
        "New Vaccinations Smoothed",
        "Total Vaccinations Per Hundred",
        "People Vaccinated Per Hundred",
        "People Fully Vaccinated Per Hundred",
        "New Vaccinations Smoothed Per Million",
        "Stringency Index",
        "Population",
        "Population Density",
        "Median Age",
        "Aged 65 Older",
        "Aged 70 Older",
        "GDP Per Capita",
        "Extreme Poverty",
        "Cardiovascular Death Rate",
        "Diabetes Prevalance",
        "Female Smokers",
        "Male Smokers",
        "Handwashing Facilities",
        "Hospital Beds Per Thousand",
        "Life Expectancy",
        "Human Development Index",
    ],
    skiprows=[0],
)

#Getting rid of one of the columns that is unneccesary
df = df[df['Location']!="World"]
```

```python
'''Zeroing out all the of columns that have NaN as their value'''

df = df.fillna(0)

df
```

# Total COVID cases vs Hospital Patients

```python
''' Based on the visualizations below, there seems to be a connection between total COVID cases and hospitalizations. The higher the total COVID cases in 
a country, the more hospitalizations there are.'''

#Creating the dataframe group to find a correlation between total cases and hospital patients. Location is also grouped for extra info
hospital_group = df.groupby(['Location', 'Total Cases'])['Hospital Patients'].max().reset_index()
#Grouping again since we only want 1 of each country instead of multiple reports from the same country. The max aggregate is taken and applied
hospital_group = hospital_group.groupby(['Location'])['Total Cases','Hospital Patients'].agg('max').reset_index()
#Only taking the values that are greater than 0 to plot
hospital_group = hospital_group[hospital_group['Hospital Patients']>0]
#Using plotly scatter plot in order to visualize the data. Also hovering and coloring the location
fig = px.scatter(hospital_group, x="Hospital Patients", y = 'Total Cases',title="Hospital Patients vs COVID Cases",hover_name='Location',
                   hover_data=['Location'],color = 'Location')
fig.show()
```

# Population vs COVID Cases

```python
'''There seems to be a clear connection between population size and covid total cases. 
The greater the population, the more cases that country/continent has.'''

#Grouping by population, location, continent in order to analyze total cases
pop_group = df.groupby(['Population', 'Location', 'Continent'])['Total Cases'].max().reset_index()
#Sorting the values based on population and location
pop_group = pop_group.sort_values(['Population', 'Location'], ascending = False)
#Plotly scatter plot to visualize each continent
fig = px.scatter(pop_group, x="Population", y="Total Cases",
           size="Population", color="Continent", hover_name="Location", facet_col = "Continent",
           log_x=True, size_max=45)

fig.show()
```

```python
#Creating a new grouped dataframe in order to work with only the columns that we are interested in
df_grouped = df.groupby(["Location", "Continent"])['Total Cases', 'Total Deaths', 'New Cases', 'ICU Patients', 'Positive Rate', 'Total Vaccinations', 'People Vaccinated', 'People Fully Vaccinated', 'New Vaccinations', 'Population', 'Aged 65 Older'].max().reset_index()
#Renaming one of the columns to make it better to understand
df_grouped = df_grouped.rename(columns={'Aged 65 Older': 'Aged 65 Older Percentage'})
df_grouped
```

# Worldwide Vaccination Visualization

```python
#Using plotly's choropleth globe graph in order to display the total vaccinations in each country
trace = go.Choropleth(
            #Creating variable locations to be the countries
            locations = df_grouped['Location'],
            #Using the names of countries for mode of location
            locationmode='country names',
            #Setting the coloring variable to be total vaccinations
            z = df_grouped['Total Vaccinations'],
            #Setting the text to be country names
            text = df_grouped['Location'],
            autocolorscale =False,
            reversescale = True,
            #Certain color scale to be used
            colorscale = 'viridis',
            marker = dict(
                line = dict(
                    color = 'rgb(0,0,0)',
                    width = 0.5)
            ),
            #Creating the color bar to the right of graph
            colorbar = dict(
                #Creating the title
                title = 'Total Vaccinations',
                tickprefix = '')
        )
#Setting data to be everything above
data = [trace]
#Using plotly go function to layout the graph properly
layout = go.Layout(
    #Creating title
    title = 'Total vaccinations per country',
    geo = dict(
        showframe = True,
        showlakes = False,
        showcoastlines = True,
        projection = dict(
            type = 'natural earth'
        )
    )
)
#Creating the fig variable which is everything from above
fig = dict( data=data, layout=layout )
#Using plotly iplot to graph
iplot(fig)
```

# Worldwide COVID Cases Visualization

```python
#Using plotly's choropleth globe graph in order to display the total cases in each country
trace = go.Choropleth(
            #Creating variable locations to be the countries
            locations = df_grouped['Location'],
            #Using the names of countries for mode of location
            locationmode='country names',
            #Setting the coloring variable to be total cases
            z = df_grouped['Total Cases'],
            #Setting the text to be country names
            text = df_grouped['Location'],
            autocolorscale =False,
            reversescale = True,
            #Certain color scale to be used
            colorscale = 'viridis',
            marker = dict(
                line = dict(
                    color = 'rgb(0,0,0)',
                    width = 0.5)
            ),
            #Creating the color bar to the right of graph
            colorbar = dict(
            #Creating the title
                title = 'Total Cases',
                tickprefix = '')
        )
#Setting data to be everything above
data = [trace]
#Using plotly go function to layout the graph properly
layout = go.Layout(
    #Creating title
    title = 'Total Cases per country',
    geo = dict(
        showframe = True,
        showlakes = False,
        showcoastlines = True,
        projection = dict(
            type = 'natural earth'
        )
    )
)
#Creating the fig variable which is everything from above
fig = dict( data=data, layout=layout )
#Using plotly iplot to graph
iplot(fig)
```

# Percent of COVID Cases Based on Population

```python
'''Just another visualization to show how population affects COVID cases'''
#Creating a new column which is total cases divided by population(percent of cases)
df['Percentage of Cases'] = (df['Total Cases']/df['Population']) * 100
#Grouping by population and location to analyze percentage of cases
df_group2 = df.groupby(['Population', 'Location'])['Percentage of Cases'].max().reset_index()
#Sorting values based on population and location
df_group2 = df_group2.sort_values(['Population', 'Location'], ascending = False)
#Using plotly scatter plot to display data and coloring by location(country)
fig = px.scatter(df_group2, x="Population", y = 'Percentage of Cases',title="Countplot",hover_name='Location',
                  hover_data=['Percentage of Cases', 'Population'], color = 'Location')
fig.show()
```

# COVID Deaths Compared to COVID Cases

```python
#Grouping by location in order to analyze total cases and total deaths
df_test_group = df_grouped.groupby(['Location'])['Total Cases', 'Total Deaths'].agg('max').reset_index()
#Creating a Deaths Total Ratio which is total deaths divided by total cases
df_test_group['Deaths Total Ratio'] = df_test_group['Total Deaths'] / df_test_group['Total Cases']
#Creating scatter plot to show the country, total cases, total deaths, and ratio
fig = px.scatter(df_test_group, x="Total Deaths", y = 'Total Cases',title="Deaths Total Ratio per Country",hover_name='Location', hover_data=['Total Cases', 'Total Deaths', 'Deaths Total Ratio'], color = 'Location')
fig.show()
```

```python
'''This box and whisker plot was created to analyze the data and see if there were any outliers. It seems like Yemen stands out, with the highest deaths total ratio. This plot
shows upper fence, median, min, etc.'''
fig = px.box(df_test_group, y='Deaths Total Ratio',notched=True,hover_name='Location')
fig.show()
```

# Percent of People Aged 65 or Older vs COVID Cases and Deaths

```python
'''There doesn't seem to be a very clear correlation between countries that have a higher old age percentage and COVID deaths or cases. There may be some exceptions but
not enough evidence. Yemen still seems to stand out as having a very high percentage of deaths even though the old age population is only 2.92 percent...'''
#Grouping by location in order to analyze total cases, total deaths, and aged 65 or older percentage
df_grouped_65_viz = df_grouped.groupby(['Location'])['Total Cases', 'Total Deaths', 'Aged 65 Older Percentage'].agg('max').reset_index()
#Creating a deaths total percentage which is total deaths divided by total cases(this is the same as deaths total ratio created previously but I am getting this data from a different 
#dataframe)
df_grouped_65_viz['Deaths Total Percentage'] = (df_grouped_65_viz['Total Deaths'] / df_grouped_65_viz['Total Cases'])*100
#Scatter plot to see the data
fig = px.scatter(df_grouped_65_viz, x='Deaths Total Percentage', y = 'Aged 65 Older Percentage',title="Percentage of 65 Older vs Deaths Percentage",hover_name='Location', color = 'Location',
                   hover_data=['Location'])
fig.show()
```

```python
'''As mentioned previously, Yemen still stands out as an outlier with the highest deaths total percentage.'''
#Box and whisker plot to see the data created above differently. This is to analyze outliers and find the median or upper fence, etc.
fig = px.box(df_grouped_65_viz, y='Deaths Total Percentage',notched=True,hover_name='Location', hover_data = ['Aged 65 Older Percentage'])                                                                                                  
fig.show()
```

# Vaccination Percentage vs Population

```python
'''There seems to be really high percentages of vaccinations in very small countries with small populations. For example Gibraltar has a 
population of 33 thousand but has 188 percentage of vaccinations.... Not sure why'''
#Grouping by location to analyze total vaccinations and population
df_group4 = df_grouped.groupby(['Location'])['Total Vaccinations', 'Population'].agg('max').reset_index()
#Creating a percent of vaccinations column which is total vaccinations divided by population
df_group4['Percent of Vaccinations'] = (df_group4['Total Vaccinations']/df_group4['Population'])*100
#Scatter plot to visualize the data
fig = px.scatter(df_group4, x='Percent of Vaccinations', y = 'Population',title="Percent of Vaccinations per Country",hover_name='Location', color = 'Location',
                   hover_data=['Location'])
fig.show()

```
